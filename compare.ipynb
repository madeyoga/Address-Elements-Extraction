```python
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
df1 = pd.read_csv('sp2-60.csv')
df2 = pd.read_csv('sp2-70.csv')
```

```python
accuracy_score(df1["POI/street"], df2["POI/street"])
```

```python
df1_np = df1["POI/street"].values
df2_np = df2["POI/street"].values
```

```python
counter = 0
for row1, row2 in zip(df1_np, df2_np):
    if row1 != row2:
        print(row1, ":::", row2)
        counter += 1
        
counter
```
